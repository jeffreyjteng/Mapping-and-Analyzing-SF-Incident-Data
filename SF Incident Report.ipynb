```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Police_Department_Incident_Reports__2018_to_Present.csv')
```

```python
df
```

```python
df=df.dropna(subset=['Latitude', 'Longitude','Incident Category','Analysis Neighborhood'])
```

```python
df.columns
```

```python
df = df.drop(['CAD Number','Neighborhoods','CNN','Supervisor District 2012','Supervisor District','Current Supervisor Districts', 'Current Police Districts','Point','HSOC Zones as of 2018-06-05','Invest In Neighborhoods (IIN) Areas','ESNCAG - Boundary File','Central Market/Tenderloin Boundary Polygon - Updated','Civic Center Harm Reduction Project Boundary'],axis=1)
```

```python
df.isna().sum()
```

```python
df['Filed Online'].unique()
```

```python
df['Filed Online'].fillna(False, inplace=True)
```

```python
def convert_datetime(dt_str):
    dt_obj = datetime.strptime(dt_str, '%Y/%m/%d %I:%M:%S %p')
    return dt_obj.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df['Incident Datetime'] = df['Incident Datetime'].apply(convert_datetime)
df['Report Datetime'] = df['Report Datetime'].apply(convert_datetime)
```

```python
df['Incident Date'] = df['Incident Date'].apply(lambda x: datetime.strptime(x, '%Y/%m/%d'))
```

```python
df
```

```python
df.to_csv("SF_Incident_Report_2018-2023.csv",index=False)
```
